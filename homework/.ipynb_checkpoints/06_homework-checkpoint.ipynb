```python
import pyspark
from pyspark.sql import SparkSession
```

```python
import os
os.environ['JAVA_HOME'] = '/usr/lib/jvm/java-17-openjdk-amd64'
```

```python
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('test') \
    .getOrCreate()
```

```python
print(spark.version)
```

```python
!wget https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2025-11.parquet
```

```python
df = spark.read.parquet('yellow_tripdata_2025-11.parquet')
```

```python
df = df.repartition(4)
```

```python
df.write.parquet('yellow/2025/11/')
```

```python
df.registerTempTable('trips_data')
```

```python
spark.sql("""
SELECT
    count(1) as trips_count
FROM
    trips_data
WHERE date_trunc('day',tpep_pickup_datetime) = '2025-11-15'
""").show()
```

```python
spark.sql("""
SELECT
    MAX(TIMESTAMPDIFF(HOUR,tpep_pickup_datetime,tpep_dropoff_datetime)) as max_duration_in_hour
FROM
    trips_data
""").show()
```

```python
df_zone = spark.read \
    .option("header", "true") \
    .csv('taxi_zone_lookup.csv')
```

```python
df.write.parquet('zones')
```

```python
df_zones = spark.read.parquet('zones/')
```

```python
df_join = df.join(df_zones, df.PULocationID==df_zones.LocationID)
```

```python
df_join.registerTempTable('trips_data_enriched')
```

```python
spark.sql("""
SELECT
    Zone,
    count(1) as trips_count
FROM
    trips_data_enriched
GROUP BY Zone
ORDER BY count(1)
""").show()
```
